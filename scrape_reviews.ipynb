```python
from threading import Thread
import urllib
import urllib2
import pandas as pd
import bs4 as bs
import re
import csv
import requests
import math
import random
user_list = pd.read_csv('/Users/rahuldhakecha/coursesfall2017/goodreads/files/users_150_reviews_final.csv')['user_id']
# exist_users_books = pd.read_csv('/Users/rahuldhakecha/coursesfall2017/goodreads/files/top_2000_user_perspective/top_2000_users_books_reviewed.csv')[['user_id','book_id']]    
```

```python
## framing urls for different pages of 2000 users
from multiprocessing import Pool
web_urls=[]
user_cnt=0
for user in user_list[0:]:
    print(user_cnt)
    print(user)
    try:
        user_cnt+=1
        opener = urllib2.build_opener()
        opener.addheaders.append(('Cookie', "_session_id2=cf89420e8052dfe265889f00060b209b"))
        htmltext = opener.open("https://www.goodreads.com/review/list/"+str(user)).read()       
        soup = bs.BeautifulSoup(htmltext,'lxml',from_encoding="utf8") 
        for find_count in soup.find_all('a',{'class':'selectedShelf'}):
            total_reviews=re.findall(r'\d+',str(find_count.string))[0]
            tot_pages=int(math.ceil(float(total_reviews)/30.0))
        for page in range(1,tot_pages+1):
            web_urls.append("https://www.goodreads.com/review/list/"+str(user)+"?page="+str(page))
    except:
        pass
with open('/Users/rahuldhakecha/coursesfall2017/goodreads/files/top_2000_user_perspective/reviews/web_urls1.csv', 'wb') as myfile:
    wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
    wr.writerow(['url_list'])
    for ele in web_urls:
        wr.writerow(ele)        
```

```python
## access review ids for different users
from multiprocessing import Pool
web_links = list(pd.read_csv('/Users/rahuldhakecha/coursesfall2017/goodreads/files/top_2000_user_perspective/reviews/web_urls.csv')['url_list_final'])   
uni = web_links[80000:85000]
final_output = []
cnt=0
def th(ur):
    global cnt
    cnt+=1
    try:
        print(re.findall(r'\d+',re.findall('list/(.+?)?page=',ur)[0])[0])
        print(cnt)
#         print('Hello')
        result=[]
        tags=[]
        book_id=[]
        opener = urllib2.build_opener()
        opener.addheaders.append(('Cookie', "_session_id2=dc3c8699ddffa4c074c86aa4f9a80f83"))
        htmltext = opener.open(ur).read()
        soup = bs.BeautifulSoup(htmltext,'lxml',from_encoding="utf8")
        for tag in soup.find_all('tr'):
            try:
                tags.append(re.findall(r'\d+',str(tag['id']))[0])
            except:
                pass
        for url2 in soup.find_all('tr',{'class':'bookalike review'}):
            book_id.append(re.findall('/book/show/(.+?)(\.|-)',str(url2))[0][0])
#         print(len(tags))
#         print(len(book_id))
        for bid,tg in zip(book_id,tags):
#             print [ur, str(bid), str(rat)]
            result.append([re.findall(r'\d+',re.findall('list/(.+?)?page=',ur)[0])[0], str(bid), str(tg)])
        return(result)
    except:
        return

p=Pool()
final_output=p.map(th, uni)
with open('/Users/rahuldhakecha/coursesfall2017/goodreads/files/top_2000_user_perspective/reviews/user_reviews_list_80000_85000.csv', 'wb') as myfile:
    wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
    wr.writerow(['user_id','book_id','review_id'])
    for ele1 in final_output:
        for ele2 in ele1:
            wr.writerow(ele2)
```
